# Jupyter Notebook - Chi-Square Test (Simple Dataset)

We will use: Study Hours vs Pass/Fail

We want to check: Are Study Hours and Result related?

## Cell 1 - Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
import seaborn as sns
```

## Cell 2 - Create Simple Dataset

```python
data = {
    "Study": ["Less","Less","Less","Less","Less",
              "More","More","More","More","More"],

    "Result": ["Fail","Fail","Fail","Pass","Fail",
               "Pass","Pass","Pass","Pass","Fail"]
}

df = pd.DataFrame(data)
df
```

## Cell 3 - Hypothesis

| Hypothesis | Meaning |
|------------|---------|
| H0 | Study and Result are independent |
| H1 | Study and Result are related |

## Cell 4 - Contingency Table (Observed Values)

```python
table = pd.crosstab(df["Study"], df["Result"])
print("Observed Table:")
print(table)
```

You will get something like:

| Study | Fail | Pass |
|-------|------|------|
| Less  | 4    | 1    |
| More  | 1    | 4    |

This is Observed Table (O).

## Cell 5 - Chi-Square Test

```python
chi2, p_value, dof, expected = chi2_contingency(table)

print("Chi-Square value:", chi2)
print("P-value:", p_value)
print("Degrees of freedom:", dof)
```

## Cell 6 - Expected Table

```python
expected_table = pd.DataFrame(expected, index=table.index, columns=table.columns)

print("Expected Table:")
print(expected_table)
```

This is Expected Table (E).

## Cell 7 - Decision

```python
alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis")
    print("Conclusion: Study and Result are related")
else:
    print("Fail to Reject Null Hypothesis")
    print("Conclusion: Study and Result are independent")
```

## Cell 8 - Bar Chart

```python
table.plot(kind="bar")
plt.title("Study vs Result")
plt.xlabel("Study Hours")
plt.ylabel("Count")
plt.show()
```

## Cell 9 - Heatmap

```python
sns.heatmap(table, annot=True, cmap="coolwarm")
plt.title("Contingency Table Heatmap")
plt.show()
```

We test: Is passing dependent on study hours?

Chi-square tells whether the relationship is real or by chance.
