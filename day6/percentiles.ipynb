```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student_marks.csv")

Q1 = df["Marks"].quantile(0.25)
Q3 = df["Marks"].quantile(0.75)

IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("Lower Limit:", lower)
print("Upper Limit:", upper)

outliers = df[(df["Marks"] < lower) | (df["Marks"] > upper)]
print("Outliers:")
print(outliers)
# Load dataset

# Box plot
plt.boxplot(df["Marks"])
plt.title("Box Plot of Students Marks")
plt.ylabel("Marks")
plt.show()
```

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("customer_spending.csv")

spending = df["Spending"]

# Percentiles
p10 = np.percentile(spending, 10)
p25 = np.percentile(spending, 25)
p30 = np.percentile(spending, 30)
p50 = np.percentile(spending, 50)
p75 = np.percentile(spending, 75)
p80 = np.percentile(spending, 80)
p90 = np.percentile(spending, 90)

print("10th percentile:", p10)
print("25th percentile (Q1):", p25)
print("50th percentile (Median):", p50)
print("75th percentile (Q3):", p75)
print("80th percentile:", p80)
print("90th percentile:", p90)

# 1 Top 10% customers
top_10 = df[df["Spending"] > p90]
print("\nTop 10% customers:")
print(top_10)

# 2 Bottom 25% customers
bottom_25 = df[df["Spending"] < p25]
print("\nBottom 25% customers:")
print(bottom_25)

# 3 Middle 50% range
print("\nMiddle 50% spending range:", p25, "to", p75)

# 4 Customers above 75th percentile
above_75 = df[df["Spending"] > p75]
print("\nCustomers above 75th percentile:", len(above_75))

# 5 Customers below 25th percentile
below_25 = df[df["Spending"] < p25]
print("Customers below 25th percentile:", len(below_25))

# 6 Is 5000 above 80th percentile?
print("\nIs 5000 above 80th percentile?", 5000 > p80)

# 7 Is 2000 below 30th percentile?
print("Is 2000 below 30th percentile?", 2000 < p30)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("student_marks.csv")

# Mean
mean_marks = df["Marks"].mean()
print("Mean:", mean_marks)

# Median
median_marks = df["Marks"].median()
print("Median:", median_marks)

# Mode
mode_marks = df["Marks"].mode()[0]
print("Mode:", mode_marks)

# Variance
variance_marks = df["Marks"].var()
print("Variance:", variance_marks)

# Standard Deviation
std_marks = df["Marks"].std()
print("Standard Deviation:", std_marks)

# Quartiles
Q1 = df["Marks"].quantile(0.25)
Q2 = df["Marks"].quantile(0.50)
Q3 = df["Marks"].quantile(0.75)

print("Q1:", Q1)
print("Q2:", Q2)
print("Q3:", Q3)

# IQR
IQR = Q3 - Q1
print("IQR:", IQR)

# Outliers
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = df[(df["Marks"] < lower_limit) | (df["Marks"] > upper_limit)]
print("Outliers:")
print(outliers)

# Box Plot
plt.boxplot(df["Marks"])
plt.title("Box Plot of Marks")
plt.show()

# Top 10% students
top_10 = df[df["Marks"] >= df["Marks"].quantile(0.90)]
print("Top 10% Students:")
print(top_10)

# Bottom 10% students
bottom_10 = df[df["Marks"] <= df["Marks"].quantile(0.10)]
print("Bottom 10% Students:")
print(bottom_10)
```
