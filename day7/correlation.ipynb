```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    "Study_Hours": [2,4,6,8,10],
    "Marks": [40,50,65,80,95],
  
}

df = pd.DataFrame(data)

# Pearson Correlation
print("\nPearson Correlation:")
print(df.corr(method='pearson'))

# Spearman Correlation
print("\nSpearman Correlation:")
print(df.corr(method='spearman'))

# Heatmap
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv("age_vs_income.csv")

# Step 2: Display first 5 rows
print("First 5 rows of dataset:")
print(df.head())

# Step 3: Find Pearson Correlation
pearson_corr = df["Age"].corr(df["Income"])
print("\nPearson Correlation:", pearson_corr)

# Step 4: Find Spearman Correlation
spearman_corr = df["Age"].corr(df["Income"], method='spearman')
print("Spearman Correlation:", spearman_corr)

# Step 5: Check Positive or Negative
if pearson_corr > 0:
    print("Correlation is Positive")
else:
    print("Correlation is Negative")

# Step 6: Check Strong or Weak
if abs(pearson_corr) > 0.7:
    print("Correlation is Strong")
elif abs(pearson_corr) > 0.4:
    print("Correlation is Moderate")
else:
    print("Correlation is Weak")

# Step 7: Scatter Plot
plt.figure()
sns.scatterplot(x="Age", y="Income", data=df)
plt.title("Scatter Plot: Age vs Income")
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()

# Step 8: Check Linear or Non-linear
difference = abs(pearson_corr - spearman_corr)

if difference < 0.1:
    print("Relationship is Linear")
else:
    print("Relationship is Non-linear")

# Step 9: Heatmap
plt.figure()
sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()

# Step 10: Which correlation is higher
if pearson_corr > spearman_corr:
    print("Pearson correlation is higher")
else:
    print("Spearman correlation is higher")

# Step 11: Outliers present or not
if difference > 0.2:
    print("Outliers are present in dataset")
else:
    print("No major outliers in dataset")

# Step 12: Which method is better
if difference < 0.1:
    print("Pearson method is better for this dataset")
else:
    print("Spearman method is better for this dataset")
```

```python
import pandas as pd

# Step 1: Load dataset
df = pd.read_csv("StudentsPerformance.csv")

# Step 2: Select required columns
scores = df[["math score", "reading score", "writing score"]]

# Step 3: Correlation matrix
corr_matrix = scores.corr()
print("Correlation Matrix:")
print(corr_matrix)

# Step 4: Individual correlations
math_reading = corr_matrix.loc["math score", "reading score"]
math_writing = corr_matrix.loc["math score", "writing score"]
reading_writing = corr_matrix.loc["reading score", "writing score"]

print("\nCorrelation between Math & Reading:", math_reading)
print("Correlation between Math & Writing:", math_writing)
print("Correlation between Reading & Writing:", reading_writing)

# Step 5: Highest and Lowest correlation
correlations = {
    "Math & Reading": math_reading,
    "Math & Writing": math_writing,
    "Reading & Writing": reading_writing
}

highest = max(correlations, key=correlations.get)
lowest = min(correlations, key=correlations.get)

print("\nHighest Correlation:", highest, "=", correlations[highest])
print("Lowest Correlation:", lowest, "=", correlations[lowest])

# Step 6: Positive or Negative
for pair, value in correlations.items():
    if value > 0:
        print(pair, ": Positive correlation")
    else:
        print(pair, ": Negative correlation")

# Step 7: Strong or Weak
for pair, value in correlations.items():
    if abs(value) > 0.7:
        print(pair, ": Strong correlation")
    elif abs(value) > 0.4:
        print(pair, ": Moderate correlation")
    else:
        print(pair, ": Weak correlation")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv("StudentsPerformance.csv")

# Step 2: Select score columns
scores = df[["math score", "reading score", "writing score"]]

# Step 3: Correlation matrix
corr = scores.corr()

# Step 4: Plot heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Step 5: Find highest and lowest correlation
math_reading = corr.loc["math score", "reading score"]
math_writing = corr.loc["math score", "writing score"]
reading_writing = corr.loc["reading score", "writing score"]

correlations = {
    "Math & Reading": math_reading,
    "Math & Writing": math_writing,
    "Reading & Writing": reading_writing
}

highest = max(correlations, key=correlations.get)
lowest = min(correlations, key=correlations.get)

print("Highly correlated subjects (Dark color):", highest)
print("Less correlated subjects (Light color):", lowest)

# Step 6: Print insights
print("\nInsights:")
print("1. Reading and Writing scores are highly correlated.")
print("2. Math has slightly lower correlation with Reading and Writing.")
print("3. All subjects are positively correlated, meaning if one score increases, others also increase.")
```
