## ❓ Question 1
### A dice is rolled. What is the probability of getting an even number?

💡 **Explanation:**
- Total outcomes = 6  →  {1, 2, 3, 4, 5, 6}
- Even numbers = {2, 4, 6}  →  3 outcomes
- Formula:  P = Favorable / Total = 3 / 6 = 0.5

```python
# Question 1 - Even number on dice

total_outcomes = 6
favorable_outcomes = 3    # {2, 4, 6}

probability = favorable_outcomes / total_outcomes

print("Total outcomes   =", total_outcomes)
print("Favorable (even) =", favorable_outcomes)
print("Probability      =", probability)
```

## ❓ Question 2
### A coin is tossed. What is the probability of getting Tail?

💡 **Explanation:**
- Sample space = {Head, Tail}  →  Total = 2
- Favorable = {Tail}  →  1
- P(Tail) = 1 / 2 = 0.5

```python
# Question 2 - Tail on coin toss

total_outcomes = 2       # Head, Tail
favorable_outcomes = 1   # only Tail

probability = favorable_outcomes / total_outcomes

print("Total outcomes    =", total_outcomes)
print("Favorable (Tail)  =", favorable_outcomes)
print("P(Tail)           =", probability)
```

## ❓ Question 3
### A dice is rolled. What is the probability of getting a number greater than 4?

💡 **Explanation:**
- Total outcomes = 6
- Numbers greater than 4 = {5, 6}  →  2 outcomes
- P = 2 / 6 = 0.333

```python
# Question 3 - Number greater than 4

total_outcomes = 6
favorable_outcomes = 2    # {5, 6}

probability = favorable_outcomes / total_outcomes

print("Total outcomes        =", total_outcomes)
print("Favorable (5 and 6)   =", favorable_outcomes)
print("P(number > 4)         =", probability)
print("P(number > 4) approx  =", round(probability, 3))
```

## ❓ Question 4
### P(Rain) = 0.65. What is P(No Rain)?

💡 **Explanation:**
- This is Complementary Event
- Formula:  P(A') = 1 - P(A)
- P(No Rain) = 1 - 0.65 = 0.35

```python
# Question 4 - Complementary Event

p_rain = 0.65

# Formula: P(No Rain) = 1 - P(Rain)
p_no_rain = 1 - p_rain

print("P(Rain)    =", p_rain)
print("P(No Rain) = 1 -", p_rain, "=", p_no_rain)

# Check: both must add up to 1
print("Check: ", p_rain, "+", p_no_rain, "=", p_rain + p_no_rain)
```

## ❓ Question 5

💡 **Explanation:**

- Total outcomes = 6
- Favorable = {3}  →  1

## ❓ Question 6
### A coin is tossed 10 times. Head comes 6 times. What is the Empirical Probability of Head?

💡 **Explanation:**
- Empirical Probability = based on real experiment
- Formula:  P = Number of times event happened / Total trials
- P(Head) = 6 / 10 = 0.6

```python
# Question 6 - Empirical Probability

total_trials = 10
head_count = 6

# Empirical Formula
p_head = head_count / total_trials

print("Total trials         =", total_trials)
print("Head appeared        =", head_count, "times")
print("Empirical P(Head)    =", p_head)

# Compare with theoretical
p_theoretical = 1 / 2
print("Theoretical P(Head)  =", p_theoretical)
print("Are they same?       =", p_head == p_theoretical)
```

## ❓ Question 8
### Dice is rolled. A = getting 2, B = getting 5. Find P(A or B).

💡 **Explanation:**
- A and B cannot happen at same time → Mutually Exclusive
- Formula:  P(A or B) = P(A) + P(B)
- P(A) = 1/6,  P(B) = 1/6
- P(A or B) = 1/6 + 1/6 = 2/6 = 0.333

```python
# Question 8 - Addition Rule (Mutually Exclusive)

total_outcomes = 6

# Event A = getting 2
p_A = 1 / total_outcomes

# Event B = getting 5
p_B = 1 / total_outcomes

# A and B cannot happen together (mutually exclusive)
# Formula: P(A or B) = P(A) + P(B)
p_A_or_B = p_A + p_B

print("P(A) = P(getting 2) =", p_A)
print("P(B) = P(getting 5) =", p_B)
print("P(A or B)           =", p_A, "+", p_B, "=", p_A_or_B)
print("P(A or B) approx    =", round(p_A_or_B, 4))
```

## ❓ Question 9
### Card drawn from deck. A = Heart, B = King. Find P(A or B).

💡 **Explanation:**
- A and B CAN happen together (King of Heart exists) → NOT Mutually Exclusive
- Formula:  P(A or B) = P(A) + P(B) - P(A and B)
- P(A) = 13/52,  P(B) = 4/52,  P(A and B) = 1/52
- P(A or B) = 13/52 + 4/52 - 1/52 = 16/52

```python
# Question 9 - Addition Rule (Not Mutually Exclusive)

total_cards = 52

# Event A = Heart card
heart_cards = 13
p_A = heart_cards / total_cards

# Event B = King card
king_cards = 4
p_B = king_cards / total_cards

# A and B together = King of Heart (only 1 card)
p_A_and_B = 1 / total_cards

# Formula: P(A or B) = P(A) + P(B) - P(A and B)
p_A_or_B = p_A + p_B - p_A_and_B

print("P(A) = P(Heart) =", heart_cards, "/", total_cards, "=", p_A)
print("P(B) = P(King)  =", king_cards, "/", total_cards, "=", p_B)
print("P(A and B)      = 1 /", total_cards, "=", p_A_and_B)
print("")
print("P(A or B) = P(A) + P(B) - P(A and B)")
print("P(A or B) =", p_A, "+", p_B, "-", p_A_and_B)
print("P(A or B) =", p_A_or_B)
print("P(A or B) = 16/52 =", round(p_A_or_B, 4))
```

## ❓ Question 10
### Check the 3 Axioms of Probability using a coin toss.

💡 **Explanation:**
- Axiom 1: P(A) >= 0  →  No probability is negative
- Axiom 2: P(S) = 1   →  All probabilities add up to 1
- Axiom 3: P(Head or Tail) = P(Head) + P(Tail)  →  for mutually exclusive

```python
# Question 10 - 3 Axioms of Probability

p_head = 0.5
p_tail = 0.5

print("=== Coin Toss ===")
print("P(Head) =", p_head)
print("P(Tail) =", p_tail)
print("")

# Axiom 1: Probability is always >= 0
print("--- Axiom 1: P(A) >= 0 ---")
print("P(Head) >= 0 :", p_head >= 0)
print("P(Tail) >= 0 :", p_tail >= 0)
print("")

# Axiom 2: Sum of all probabilities = 1
print("--- Axiom 2: P(S) = 1 ---")
total = p_head + p_tail
print("P(Head) + P(Tail) =", p_head, "+", p_tail, "=", total)
print("Sum = 1 ?", total == 1)
print("")

# Axiom 3: Mutually exclusive events
print("--- Axiom 3: P(Head or Tail) = P(Head) + P(Tail) ---")
# Head and Tail cannot happen together
p_head_or_tail = p_head + p_tail
print("P(Head or Tail) =", p_head_or_tail)
print("")

print("=== All 3 Axioms Verified! ===")
```
