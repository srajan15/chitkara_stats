# Module 1 – Basics of Probability

## 1. What is Probability
Probability means chance of happening an event.
Probability value is always between 0 and 1.

- 0 → Impossible event
- 1 → Certain event
- Between 0 and 1 → Possible event

```python
print("Probability of sun rising =", 1)
print("Probability of getting 7 on dice =", 0)
print("Probability of getting Head =", 0.5)
```

## 2. Random Experiment
An experiment where result is not known in advance.

```python
import random

coin = random.choice(["Head", "Tail"])
print("Coin Toss Result:", coin)

dice = random.randint(1, 6)
print("Dice Roll Result:", dice)
```

## 3. Sample Space
Sample space means all possible outcomes.

```python
coin_space = ["H", "T"]
dice_space = [1,2,3,4,5,6]
two_coin_space = ["HH", "HT", "TH", "TT"]

print("Coin Sample Space:", coin_space)
print("Dice Sample Space:", dice_space)
print("Two Coin Sample Space:", two_coin_space)
```

## 4. Event
Event is a subset of sample space.

```python
sample_space = [1,2,3,4,5,6]
event_A = [2,4,6]

print("Sample Space:", sample_space)
print("Event A (Even numbers):", event_A)
```

## 5. Complementary Event
Formula: P(A') = 1 - P(A)

```python
P_head = 0.5
P_not_head = 1 - P_head

print("P(Head):", P_head)
print("P(Not Head):", P_not_head)
```

## 6. Probability Formula
P(A) = Favorable Outcomes / Total Outcomes

```python
favorable = 3
total = 6

probability = favorable / total
print("Probability of even number:", probability)
```

## 7. Classical Probability
Example: Deck of cards

```python
total_cards = 52
heart_cards = 13

probability = heart_cards / total_cards
print("Probability of Heart:", probability)
```

## 8. Empirical Probability
Based on real experiment data

```python
results = ["H","T","H","H","T","H","T","H","H","T"]

heads = results.count("H")
total = len(results)

probability = heads / total
print("Experimental Probability of Head:", probability)
```

## 10. Addition Rule (Mutually Exclusive) of Probability

**Given:**
- Dice experiment
- Event A = Getting number 2
- Event B = Getting number 5

**Sample space of dice:** $S = \{1, 2, 3, 4, 5, 6\}$

---

### Step 1 – Find P(A)

Event A = getting 2 → Only one favorable outcome: $\{2\}$

$$P(A) = \frac{1}{6}$$

---

### Step 2 – Find P(B)

Event B = getting 5 → Only one favorable outcome: $\{5\}$

$$P(B) = \frac{1}{6}$$

---

### Step 3 – Use Addition Rule

We want the probability of getting **2 OR 5**.

Since both cannot happen together in one dice roll, they are **mutually exclusive events**.

So the formula becomes:

$$P(A \cup B) = P(A) + P(B)$$

$$P(A \cup B) = \frac{1}{6} + \frac{1}{6} = \frac{2}{6} = \frac{1}{3}$$

---

### ✅ Final Answer

$$P(\text{2 or 5}) = \frac{1}{3}$$

---

**Simple Meaning:**
Out of 6 possible outcomes, 2 outcomes are favorable (2 and 5), so probability = 2/6 = **1/3**.

```python
P_A = 1/6
P_B = 1/6

P_union = P_A + P_B
print("P(A ∪ B) =", P_union)
```

## Addition Rule (Not Mutually Exclusive) – Explanation

### Situation

We pick one card from a deck of 52 cards.

- **Event A** = Getting a Heart
- **Event B** = Getting a King

> ⚠️ There is one card which is **both** Heart and King → **King of Hearts**
> Since events can happen together, they are **NOT mutually exclusive.**

---

### Step 1 – Find P(A)

Hearts = 13 cards

$$P(A) = \frac{13}{52}$$

---

### Step 2 – Find P(B)

Kings = 4 cards

$$P(B) = \frac{4}{52}$$

---

### Step 3 – Find P(A ∩ B)

Heart **AND** King together → King of Hearts = 1 card

$$P(A \cap B) = \frac{1}{52}$$

---

### Step 4 – Apply Addition Rule Formula

$$P(A \cup B) = P(A) + P(B) - P(A \cap B)$$

$$P(A \cup B) = \frac{13}{52} + \frac{4}{52} - \frac{1}{52} = \frac{16}{52} = \frac{4}{13}$$

---

### Why We Subtract?

Because **King of Hearts** is counted **two times:**

- Once in Hearts
- Once in Kings

So we subtract it **once** to correct the double-counting.

---

### Table Summary

| Card Type       | Count |
|-----------------|-------|
| Hearts          | 13    |
| Kings           | 4     |
| King of Hearts  | 1     |

**Total favorable outcomes:**

$$13 + 4 - 1 = 16$$

$$P = \frac{16}{52}$$

---

### ✅ Final Answer

$$P(\text{Heart or King}) = \frac{16}{52} = \frac{4}{13}$$

```python
P_A = 13/52
P_B = 4/52
P_intersection = 1/52

P_union = P_A + P_B - P_intersection
print("P(A ∪ B) =", P_union)
```
