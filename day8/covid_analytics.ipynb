# 🦠 COVID-19 Data Analytics for Beginners
### Topics Covered: Mean, Mode, Median, Std, Variance, Percentile, Quantile, Correlation

## Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print('✅ All libraries imported successfully!')
```

## Step 2: Load the COVID Dataset

```python

df = pd.read_csv("covid_data.csv")

print('✅ Dataset loaded!')
print(f'Shape: {df.shape[0]} rows x {df.shape[1]} columns')
df.head()
```

## Step 3: Basic Dataset Info

```python
# See column names and data types
print('📋 Column Names:')
print(df.columns.tolist())

print('\n📊 Data Types:')
print(df.dtypes)

print('\n❓ Missing Values:')
print(df.isnull().sum())
```

## Step 4: 📐 MEAN — Average Value

```python
# Mean = Sum of all values / Number of values
# It tells us the 'average' number

mean_confirmed = df['Confirmed'].mean()
mean_deaths    = df['Deaths'].mean()
mean_recovered = df['Recovered'].mean()
mean_active    = df['Active'].mean()

print('📐 MEAN (Average)')
print('-' * 35)
print(f'Mean Confirmed Cases : {mean_confirmed:,.2f}')
print(f'Mean Deaths          : {mean_deaths:,.2f}')
print(f'Mean Recovered       : {mean_recovered:,.2f}')
print(f'Mean Active Cases    : {mean_active:,.2f}')
```

## Step 5: 🔢 MEDIAN — Middle Value

```python
# Median = The middle value when data is sorted
# It is better than mean when data has extreme outliers

median_confirmed = df['Confirmed'].median()
median_deaths    = df['Deaths'].median()
median_recovered = df['Recovered'].median()
median_active    = df['Active'].median()

print('🔢 MEDIAN (Middle Value)')
print('-' * 35)
print(f'Median Confirmed Cases : {median_confirmed:,.2f}')
print(f'Median Deaths          : {median_deaths:,.2f}')
print(f'Median Recovered       : {median_recovered:,.2f}')
print(f'Median Active Cases    : {median_active:,.2f}')
```

## Step 6: 🎯 MODE — Most Frequent Value

```python
# Mode = The value that appears most often
# Very useful for categorical data like WHO Region

mode_region = df['WHO Region'].mode()[0]
mode_new_deaths = df['New deaths'].mode()[0]

print('🎯 MODE (Most Frequent Value)')
print('-' * 35)
print(f'Most common WHO Region  : {mode_region}')
print(f'Mode of New Deaths      : {mode_new_deaths}')

print('\n📊 WHO Region Frequency Count:')
print(df['WHO Region'].value_counts())
```

## Step 7: 📏 STANDARD DEVIATION — Spread of Data

```python
# Standard Deviation = How spread out values are from the mean
# High std = data is very spread out
# Low std  = data is close to the mean

std_confirmed = df['Confirmed'].std()
std_deaths    = df['Deaths'].std()
std_recovered = df['Recovered'].std()
std_active    = df['Active'].std()

print('📏 STANDARD DEVIATION')
print('-' * 35)
print(f'Std of Confirmed Cases : {std_confirmed:,.2f}')
print(f'Std of Deaths          : {std_deaths:,.2f}')
print(f'Std of Recovered       : {std_recovered:,.2f}')
print(f'Std of Active Cases    : {std_active:,.2f}')
```

## Step 8: 📦 VARIANCE — Squared Spread

```python
# Variance = Standard Deviation squared
# It shows how much values differ from the mean

var_confirmed = df['Confirmed'].var()
var_deaths    = df['Deaths'].var()
var_recovered = df['Recovered'].var()
var_active    = df['Active'].var()

print('📦 VARIANCE')
print('-' * 35)
print(f'Variance of Confirmed Cases : {var_confirmed:,.2f}')
print(f'Variance of Deaths          : {var_deaths:,.2f}')
print(f'Variance of Recovered       : {var_recovered:,.2f}')
print(f'Variance of Active Cases    : {var_active:,.2f}')
```

## Step 9: 📊 PERCENTILE — Position in Data

```python
# Percentile = % of data that falls BELOW a value
# Example: 90th percentile means 90% of countries have fewer cases

p25 = np.percentile(df['Confirmed'], 25)
p50 = np.percentile(df['Confirmed'], 50)   # same as median
p75 = np.percentile(df['Confirmed'], 75)
p90 = np.percentile(df['Confirmed'], 90)

print('📊 PERCENTILE — Confirmed Cases')
print('-' * 38)
print(f'25th Percentile (P25) : {p25:,.0f}')
print(f'50th Percentile (P50) : {p50:,.0f}  ← same as Median')
print(f'75th Percentile (P75) : {p75:,.0f}')
print(f'90th Percentile (P90) : {p90:,.0f}')

# Deaths percentile
print('\n📊 PERCENTILE — Deaths')
print('-' * 38)
for p in [25, 50, 75, 90]:
    val = np.percentile(df['Deaths'], p)
    print(f'{p}th Percentile : {val:,.0f}')
```

## Step 10: 📦 QUANTILE — Split Data into Equal Parts

```python
# Quantile divides data into equal groups
# Q1=0.25, Q2=0.50 (median), Q3=0.75

print('📦 QUANTILE — Confirmed Cases')
print('-' * 38)

q1 = df['Confirmed'].quantile(0.25)
q2 = df['Confirmed'].quantile(0.50)
q3 = df['Confirmed'].quantile(0.75)

print(f'Q1 (25%) : {q1:,.0f}')
print(f'Q2 (50%) : {q2:,.0f}  ← Median')
print(f'Q3 (75%) : {q3:,.0f}')
print(f'IQR (Q3 - Q1) : {q3 - q1:,.0f}  ← Interquartile Range')

print('\n📦 QUANTILE — Deaths')
print('-' * 38)
q_deaths = df['Deaths'].quantile([0.25, 0.50, 0.75])
print(q_deaths)
```

## Step 11: 🔗 CORRELATION — Relationship Between Columns

```python
# Correlation = How strongly two columns are related
# Value between -1 and +1
# +1 = perfect positive relationship
# -1 = perfect negative relationship
#  0 = no relationship

# Select only number columns
num_cols = ['Confirmed', 'Deaths', 'Recovered', 'Active', 'New cases', 'New deaths']

correlation_matrix = df[num_cols].corr()

print('🔗 CORRELATION MATRIX')
print('-' * 60)
print(correlation_matrix.round(2))

print('\n💡 Key Insight:')
corr_conf_deaths = df['Confirmed'].corr(df['Deaths'])
print(f'Correlation between Confirmed & Deaths = {corr_conf_deaths:.4f}')
if corr_conf_deaths > 0.7:
    print('→ Strong Positive Correlation: More confirmed cases = More deaths')
```

## Step 12: 📋 Full Summary Statistics

```python
# describe() gives you count, mean, std, min, percentiles, max — all at once!
print('📋 Full Summary Statistics:')
df[['Confirmed', 'Deaths', 'Recovered', 'Active']].describe().round(2)
```

## Step 13: 📊 Visualizations

```python
# --- Plot 1: Bar Chart — Top 10 Countries by Confirmed Cases ---
plt.figure(figsize=(12, 5))
top10 = df.nlargest(10, 'Confirmed')
plt.bar(top10['Country/Region'], top10['Confirmed'], color='steelblue')
plt.title('Top 10 Countries by Confirmed COVID-19 Cases', fontsize=14)
plt.xlabel('Country')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- Plot 2: Correlation Heatmap ---
plt.figure(figsize=(9, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap — COVID-19 Columns', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# --- Plot 3: Boxplot to show spread (std, variance, percentile visually) ---
plt.figure(figsize=(10, 5))
df[['Confirmed', 'Deaths', 'Recovered', 'Active']].boxplot()
plt.title('Boxplot — Spread of COVID-19 Numbers', fontsize=14)
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# --- Plot 4: Histogram — Distribution of Deaths ---
plt.figure(figsize=(8, 4))
plt.hist(df['Deaths'], bins=6, color='tomato', edgecolor='black')
plt.axvline(df['Deaths'].mean(),   color='blue',  linestyle='--', label=f"Mean = {df['Deaths'].mean():,.0f}")
plt.axvline(df['Deaths'].median(), color='green', linestyle='--', label=f"Median = {df['Deaths'].median():,.0f}")
plt.title('Distribution of Deaths per Country', fontsize=14)
plt.xlabel('Deaths')
plt.ylabel('Number of Countries')
plt.legend()
plt.tight_layout()
plt.show()
```

## ✅ Summary Table — All Statistics

```python
# Quick Reference Summary
cols = ['Confirmed', 'Deaths', 'Recovered', 'Active']

summary = pd.DataFrame({
    'Mean'     : df[cols].mean().round(2),
    'Median'   : df[cols].median().round(2),
    'Mode'     : [df[c].mode()[0] for c in cols],
    'Std Dev'  : df[cols].std().round(2),
    'Variance' : df[cols].var().round(2),
    'P25'      : df[cols].quantile(0.25).round(2),
    'P75'      : df[cols].quantile(0.75).round(2),
})

print('✅ COMPLETE STATISTICS SUMMARY')
print('=' * 60)
summary
```
